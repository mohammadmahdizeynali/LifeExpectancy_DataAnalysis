```python
# import library that we need

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import dataframe

df = pd.read_csv("primary_dataset.csv")
```

```python
df = df.groupby(['Country', 'Region']).mean().drop(columns='Year').reset_index()
df.head()
```

```python
sorted_life_df = df.sort_values(by='Life_expectancy', ascending=False)
max_life_country = sorted_life_df.head(10)['Country'].tolist()
max_life_country
```

```python
# List of 10 specific countries for which you want to export the data
max_life_country_list = ['Japan',
 'Switzerland',
 'Iceland',
 'Italy',
 'Australia',
 'Spain',
 'Sweden',
 'France',
 'Canada',
 'Iran, Islamic Rep.']

# Filter the DataFrame to include data only for the 10 specific countries
max_life_country_df = df[df['Country'].isin(max_life_country_list)].sort_values(by='Life_expectancy', ascending=False)
max_life_country_df.head(10)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(max_life_country_df['Country'], max_life_country_df['Life_expectancy'])
plt.xlabel('Country')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy (70-85) for Ten Countries')
plt.ylim(70, 85)  # Set the y-axis limits to show only the range of 60-80
plt.xticks(rotation=45)
plt.show()
```

```python
# now we want to plot diagram of importance features
importance_features = ['Infant_deaths', 'Under_five_deaths','Adult_mortality', 'GDP_per_capita', 'Alcohol_consumption', 'Population_mln']

def plot_features(feature):
    plt.figure(figsize=(12, 6))
    plt.bar(max_life_country_df['Country'], max_life_country_df[feature])
    plt.xlabel('Country')
    plt.ylabel(feature)
    plt.title(f'{feature} for Ten Countries')
    plt.xticks(rotation=45)
    plt.show()

for feature in importance_features:
    plot_features(feature)
    
```
