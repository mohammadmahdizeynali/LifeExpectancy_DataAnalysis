```python
# import library that we need

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import dataset

df = pd.read_csv("primary_dataset.csv")
df.head()
```

```python
df['Life_expectancy'].dtype
```

```python
df_grouped_by_region = df.groupby('Region')['Life_expectancy'].mean().reset_index()
df_grouped_by_region = df_grouped_by_region.rename(columns={'Life_Expectancy': 'mean_life_expectancy'})
df_grouped_by_region.head()
```

```python
# Plotting the data on a bar plot
plt.figure(figsize=(12, 6))
plt.bar(df_grouped_by_region['Region'], df_grouped_by_region['Life_expectancy'], color='skyblue')
plt.xlabel('Region')
plt.ylabel('Mean Life Expectancy')
plt.title('Mean Life Expectancy by Region')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
df_grouped = df.groupby(['Region', 'Year'])['Life_expectancy'].mean().reset_index()
df_grouped.head(20)
```

```python
plt.figure(figsize=(12, 6))
for region in df_grouped['Region'].unique():
    region_data = df_grouped[df_grouped['Region'] == region]
    plt.plot(region_data['Year'], region_data['Life_expectancy'], label=region)

plt.xlabel('Year')
plt.ylabel('Mean Life Expectancy')
plt.title('Mean Life Expectancy by Region over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```
